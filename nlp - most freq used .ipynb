```python
text="'' सनातनवाद्यांनी व प्रतिगाम्यांनी समाज रसातळाला नेला असताना या अंधारात बाळशास्त्री जांभेकर यांनी 'दर्पण'च्या माध्यमातून पहिली ज्ञानज्योत तेववली , '' असे प्रतिपादन नटसम्राट प्रभाकर पणशीकर यांनी केले . दर्पणकार बाळशास्त्री जांभेकर यांच्या १९५व्या जयंतीनिमित्त महाराष्ट्र संपादक परिषद व सिंधुदुर्ग जिल्हा मराठी पत्रकार संघाच्या वतीने तसेच महाराष्ट्र जर्नलिस्ट फाउंडेशन व महाराष्ट्र ग्रामीण पत्रकार संघाच्या सहभागाने अभिवादन कार्यक्रम आयोजित केला होता . महाराष्ट्र संपादक परिषदेचे कार्याध्यक्ष यशवंत पाध"
```

```python
fd=nltk.FreqDist(text.split())
fd
```

```python
from nltk.probability import ConditionalFreqDist
cfd=ConditionalFreqDist((len(word),word) for word in text.split())
#list of conditons
cfd.conditions()
```

```python
cfd[4]
```

```python
indiandata = indian.words(fileids = 'marathi.pos')
cdf = ConditionalFreqDist((len(word), word) for word in indiandata)
cdf[4]
```

```python
data = []
for words in indiandata:
    if len(words)>5:
        data.append(words)
```

```python
fd = nltk.FreqDist(data)
fd
```

```python
sorted_fd = sorted(fd.items(), key = lambda x:x[1])
```

```python
print(sorted_fd)
```

```python
word = list(sorted_fd)[len(sorted_fd)-1]
```
